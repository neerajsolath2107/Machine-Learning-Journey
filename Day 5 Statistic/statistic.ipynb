```python
df["Age Group"] = pd.cut(
    df["Age"],
    bins = [0,12,18,40,60,100],
    labels = ["child","Teen","Adult","Middle Age","Senior"])
```

```python
# normalize the data
df["Fare norm"] = (df["Fare"] - df["Fare"].mean())/df["Fare"].std()
```
