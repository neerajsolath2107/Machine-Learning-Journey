```python
li = [2,42,32,3,18,23]
# count odd and even number in list 
odd = 0
even = 0 
for i in li :
    if i % 2 == 0:
        even += 1
    else :
        odd += 1
print("odd : " , odd)
print("even : " ,even)
```

```python
# reverse a list without using reverse 
li2 = li[::-1]
li2
```

```python
li2.append(2)
li2
```

```python
li3 = list(set(li2))  # does not maintain order
```

```python
li3  
```

```python
# it maintains order
li4  = []    
for i in li2 :
    if i not in li4 :
        li4.append(i)
```

```python
# print fabbinoaice 
def fabinoci(n):
    a,b = 0,1
    for _ in range(n):
        print(a,end = " ")
        a, b = b, a + b
```

```python
fabinoci(7)
```
